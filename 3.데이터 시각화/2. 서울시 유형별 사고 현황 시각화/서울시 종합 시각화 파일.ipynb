## 1. 서울시 월별 개인형이동장치 사고 현황 시각화

```python
df = pd.DataFrame({
    '월별': ['1월', '2월', '3월', '4월'],
    '17년': [11, 22, 33, 44],
    '18년': [20, 25, 30, 35],
    '19년': [1, 2, 3, 4]})
df = df.set_index('월별')


```

```python
plt.figure(figsize=(8, 6)) # 그래프 크기 설정
plt.plot(df['17년'], label='17년') # 17년 그래프 추가
plt.plot(df['18년'], label='18년') # 18년 그래프 추가
plt.plot(df['19년'], label='19년') # 19년 그래프 추가
plt.legend() # 범례 추가
plt.xlabel('월별') # x축 제목 설정
plt.ylabel('사고건수') # y축 제목 설정
plt.show() # 그래프 출력
```

## 5. 서울시 구별 개인형 이동장치 사고 현황

```python
# 데이터 확인
PM_total
```

컬럼명 및 '자치구별' 컬럼을 인덱스로 재설정

```python
PM_total = PM_total[['자치구별','PM총합']]
PM_total = PM_total.set_index('자치구별')
```

자치구별 종합 사고건수에 관한 파이차트 그리기

```python
#각 자치구 지정을 위한 라벨 생성
labels = list(PM_total.index)

#파이차트 생성 시 웨지의 속성을 지정
wedgeprops={'width': 2, 'edgecolor': 'w', 'linewidth': 1} #웨지 넓이:2, 엣지컬러:하얀, 선굵기:1로 설정

plt.pie(PM_total['PM총합'], labels=labels, radius=3, autopct = '%.1f%%', startangle=90, wedgeprops=wedgeprops,pctdistance=0.8)
#반지름 3, 소숫점 첫째 자리까지, 시작각도 90도, 웨지그룹 설정, 값표시 위치 설정
plt.show()
```
